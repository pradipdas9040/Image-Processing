**Logarithmic Transformation**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
   
# Read an image
image = cv2.cvtColor(cv2.imread('/content/drive/MyDrive/ImageProcessing1/image1.jpg', 1), cv2.COLOR_RGB2BGR)
   
# Apply log transformation method
c = 255 / np.log(1 + np.max(image))
log_image = c * (np.log(image + 1))
   
# Specify the data type so that
# float value will be converted to int
log_image = np.array(log_image, dtype = np.uint8)
   
fig = plt.figure(figsize=(15, 10))
rows = 1
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(image)
fig.add_subplot(rows, columns, 2)
plt.title('Logrithmic Transformation')
plt.imshow(log_image)
plt.show()
```

**Power-law Transformation**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
   
# Read an image
image = cv2.cvtColor(cv2.imread('/content/drive/MyDrive/ImageProcessing1/image1.jpg', 1), cv2.COLOR_RGB2BGR)
   
fig = plt.figure(figsize=(15, 10))
rows = 2
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(image)
fig.add_subplot(rows, columns, 2)
# Apply power-law transformation method
c = 1
gamma = 2.0
power_image = c * np.array(255*(image/255)**gamma,dtype='uint8')
# Specify the data type so that
# float value will be converted to int
power_image = np.array(power_image, dtype = np.uint8)
plt.title('Gamma = 2.0')
plt.imshow(power_image)
fig.add_subplot(rows, columns, 3)
# Apply power-law transformation method
c = 1
gamma = 0.5
power_image = c * np.array(255*(image/255)**gamma,dtype='uint8')
# Specify the data type so that
# float value will be converted to int
power_image = np.array(power_image, dtype = np.uint8)
plt.title('Gamma = 0.5')
plt.imshow(power_image)
fig.add_subplot(rows, columns, 4)
# Apply power-law transformation method
c = 1
gamma = 1
power_image = c * np.array(255*(image/255)**gamma,dtype='uint8')
# Specify the data type so that
# float value will be converted to int
power_image = np.array(power_image, dtype = np.uint8)
plt.title('Gamma = 1')
plt.imshow(power_image)
plt.show()
```

**Piecewise Linear Transformation - Contrast Stretching**

```python
import cv2
import numpy as np
  
# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, s1, r2, s2):
    if (0 <= pix and pix <= r1):
        return (s1 / r1)*pix
    elif (r1 < pix and pix <= r2):
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2
  
# Open the image.
img = cv2.imread('/content/drive/MyDrive/ImageProcessing1/image3.jpg', 0)
  
# Define parameters.
r1 = 70
s1 = 45
r2 = 140
s2 = 210
  
# Vectorize the function to apply it to each value in the Numpy array.
pixelVal_vec = np.vectorize(pixelVal)
  
# Apply contrast stretching.
contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)

# Display both images
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
fig.add_subplot(rows, columns, 2)
plt.title('Constrast Stretched Image')
plt.imshow(contrast_stretched, cmap = 'gray')
fig.add_subplot(rows, columns, 3)
plt.title('Original Image Histogram')
plt.hist(img.ravel(),256,[0,256])
fig.add_subplot(rows, columns, 4)
plt.title('Contrast Stretched Image Histogram')
plt.hist(contrast_stretched.ravel(),256,[0,256])
plt.show()
```

**Piecewise Linear Transformation - Intensity Level Slicing**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
  
# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, r2):
    if (0 <= pix and pix <= r1):
        return pix
    elif (r1 < pix and pix <= r2):
        return 100
    else:
        return pix
  
# Open the image.
img = cv2.imread('/content/drive/MyDrive/ImageProcessing1/image3.jpg', 0)
  
# Define parameters.
r1 = 70
r2 = 140
  
# Vectorize the function to apply it to each value in the Numpy array.
pixelVal_vec = np.vectorize(pixelVal)
  
# Apply contrast stretching.
intensitySliced_stretched = pixelVal_vec(img, r1, r2)

# Display both images
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
fig.add_subplot(rows, columns, 2)
plt.title('Intensity Level Sliced Image')
plt.imshow(intensitySliced_stretched, cmap = 'gray')
fig.add_subplot(rows, columns, 3)
plt.title('Original Image Histogram')
plt.hist(img.ravel(),256,[0,256])
fig.add_subplot(rows, columns, 4)
plt.title('Intensity Level Sliced Image Histogram')
plt.hist(intensitySliced_stretched.ravel(),256,[0,256])
plt.show()
```

## **SPATIAL CORRELATION AND CONVOLUTION**

```python
# Importing OpenCV and Numpy Libraries
import cv2
import numpy as np
  
# Reads image from the disk using cv2.imread() function 
img = cv2.imread('/content/drive/MyDrive/ImageProcessing1/image3.jpg')
  
# Apply identity kernel
# In an Identity Kernel matrix the value of the middle element is 1
# The values of all the other elements are 0
id_kernel = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])
  
# Filtered image is obtained using the variable flt_img
# cv2.fliter2D() is the function used
# src is the source of image(here, img)
# ddepth is destination depth. -1 will mean output image will have same depth as input image
# kernel is used for specifying the kernel operation (here, id_kernel)
flt_img = cv2.filter2D(src=img, ddepth=-1, kernel=id_kernel)
  
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
fig.add_subplot(rows, columns, 2)
plt.title('After Using Convolution Mask On Image')
plt.imshow(flt_img, cmap = 'gray')
plt.show()
```

**Box Filter**

```python
# Importing the OpenCV, Numpy and Mat libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
  
# Reading the image from the disk using cv2.imread() function
# Showing the original image using matplotlib library function plt.imshow()
img = cv2.imread('/content/drive/MyDrive/ImageProcessing1/image4.jpg')

#img = np.zeros((28,28))
#for i in range(28):
#  for j in range(28):
#    img[i][j] = (int)((random.randrange(10)*0.1)*255)
  
# Kernel for box blur filter
# It is a unity matrix which is divided by 9 
box_blur_ker = np.array([[1/9, 1/9, 1/9],
                    [1/9, 1/9, 1/9],
                    [1/9, 1/9, 1/9]])
  
# Applying Box Blur effect
# Using the cv2.filter2D() function
# src is the source of image(here, img)
# ddepth is destination depth. -1 will mean output image will have same depth as input image
# kernel is used for specifying the kernel operation (here, box_blur_ker)
Box_blur = cv2.filter2D(src=img, ddepth=-1, kernel=box_blur_ker)
  
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
fig.add_subplot(rows, columns, 2)
plt.title('Applying Box Filter')
plt.imshow(Box_blur, cmap = 'gray')
plt.show()
```

**Sharpening**

```python
# Importing the OpenCV, Numpy and Matplotlib libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Reading the image from the disk using cv2.imread() function
# Showing the original image using matplotlib library function plt.imshow()
img = cv2.imread('/content/drive/MyDrive/ImageProcessing1/image4.jpg')

# Apply kernel for sharpening
sharp_kernel = np.array([[0, -1, 0],
					[-1, 5, -1],
					[0, -1, 0]])

# Sharpeneded image is obtained using the variable sharp_img
# cv2.fliter2D() is the function used
# src is the source of image(here, img)
# ddepth is destination depth. -1 will mean output image will have same depth as input image
# kernel is used for specifying the kernel operation (here, sharp_kernel)
sharp_img = cv2.filter2D(src=img, ddepth=-1, kernel=sharp_kernel)

fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2
fig.add_subplot(rows, columns, 1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
fig.add_subplot(rows, columns, 2)
plt.title('Sharpening')
plt.imshow(sharp_img, cmap = 'gray')
plt.show()
```
