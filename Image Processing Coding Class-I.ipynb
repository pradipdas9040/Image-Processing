

**Adjacency of a pixels (4-adjacency and 8-adjacency)**







````python
#Defining a sample 2D matrix
arr=np.zeros((4,4)) 
(arr[0][3],arr[0][2],arr[1][1],arr[1][2],arr[2][2],arr[3][1]) = (1,1,1,1,1,1)
plt.imshow(arr,cmap="gray")
plt.xticks([0,1,2,3])
plt.yticks([0,1,2,3])
plt.show()

# Determining pixels that are 4-adjacent

index=[]    # Defining an empty list to stores index of pixels that are 4-adjacent

def index_update(pix_loc):  #Function to update the index list
  if pix_loc not in index:
      index.append(pix_loc)  

def colour_four_connect_pix(arr): #Function to assign a value 2 to the pixels that are 4-adjacent
    for loc in index:
      arr[loc[0]][loc[1]] = 2
    plt.imshow(arr,cmap="gray") 
    plt.xticks([0,1,2,3])
    plt.yticks([0,1,2,3])
    plt.show()

#Calculating the 4-adjacent pixels
for i in range(arr.shape[0]):
  for j in range(arr.shape[1]):
    if arr[i][j]==1:
        if i!=0 and i!= (arr.shape[0]-1) and j!=0 and j!= (arr.shape[1]-1): # For non-border pixels
          if arr[i-1][j] == 1 or arr[i+1][j] == 1 or arr[i][j-1] == 1 or arr[i][j+1] == 1: # Check for top-down and left-right neighbours of (i.j)
            index_update((i,j))
        if i==0: # For pixels in the first row
          if j==0:  # For (0,0) location
            if arr[i][j+1] == 1 or arr[i+1][j] == 1:
              index_update((i,j))
          elif j!=arr.shape[1]-1: # For intermediate locations in first row
            if arr[i][j+1] == 1 or arr[i+1][j] == 1 or arr[i][j-1] == 1:
              index_update((i,j))
          else:     # For last pixel in first row 
            if arr[i][j-1] == 1 or arr[i+1][j] == 1:
              index_update((i,j))
        if i == (arr.shape[0]-1): # For pixels in the last row
          if j == 0: # For first pixel in the last row
            if arr[i-1][j] == 1 or arr[i][j+1] == 1:
              index_update((i,j))
          elif j != arr.shape[1]-1: # For intermediate pixels in the last row
            if arr[i][j+1] == 1 or arr[i-1][j] == 1 or arr[i][j-1] == 1:
              index_update((i,j))
          else:  # For last pixel in the last row
            if arr[i-1][j] == 1 or arr[i][j-1] == 1:
              index_update((i,j))
        if (j == 0 and i != 0) and (j == 0 and i != arr.shape[0]-1): # Other than first and last pixel in the first column
          if arr[i-1][j]==1 or arr[i+1][j]==1 or arr[i][j+1] == 1:
            index_update((i,j))
        if (j == arr.shape[1]-1 and i != 0) and (j == arr.shape[1]-1 and i != arr.shape[0]-1): # Other than first and last pixel in the last column
          if arr[i-1][j]==1 or arr[i+1][j]==1 or arr[i][j-1] == 1:
            index_update((i,j))

print(index) # Print list of indices of pixels that are 4-adjacent
colour_four_connect_pix(arr)
````

**Connected Component Labelling in binary images**
(Assume 4-connectivity):

```
r: top neighbor of pixel p
t: left neighbor of pixel p
m: no. of rows
n: no. of columns
Algorithm connected_component_labelling():
{
for i := 1 to m do
	for j :=1 to n do
    if intensity(p) = 1 and (intensity(r) = intensity(t) = 0) then:
         Assign a new label to pixel p
	if intensity(p) = 1 and only one of its two neighbors are 1 then:
		Assign the label of the neighbor with an intensity of 1 to pixel p
	if intensity(p) = 1 and (intensity(r) = intensity(t) = 1) then:
		     if (label(r) = label(t)) then:
			      set label of pixel p to the label(t)
		     else:
                assign any one of the two labels to pixel p and mark both the  labels equivalent
}
```





````python
# Defining the binary image
arr = np.zeros((7,7))
arr[2][1] = arr[2][3] = arr[3][1] = arr[3][2] = arr[3][3] = arr[3][5] = arr[4][2] = arr[4][5] = arr[5][2] = arr[5][4] = arr[5][5] = arr[6][1] = arr[6][2] = 1
plt.imshow(arr)
plt.show()
# Algorithm
label = 0  # Variable to assign labels
equi_pairs=[] # List of equivalent label pairs
label_mat = np.zeros((7,7)) # 2D matrix to store labels corresponding to the (i,j)th pixel in arr

for i in range(arr.shape[0]):
  for j in range(arr.shape[1]):
    if arr[i][j] == 1 and arr[i-1][j] == 0 and arr[i][j-1] == 0: #If top and right pixel of (i,j) is 0
      label = label + 1 
      label_mat[i][j] = label     # Assign a new label to (i,j)
    if (arr[i][j] == 1 and arr[i-1][j] == 1 and arr[i][j-1] == 0) or (arr[i][j] == 1 and arr[i-1][j] == 0 and arr[i][j-1] == 1): #If either of the top and right pixel of (i,j) is 1
      if arr[i-1][j] == 1:
        label_mat[i][j] = label_mat[i-1][j]
      else:
        label_mat[i][j] = label_mat[i][j-1]
    if arr[i][j] == 1 and arr[i-1][j] == 1 and arr[i][j-1] == 1: #If both top and right pixel of (i,j) is 1
      if label_mat[i-1][j] == label_mat[i][j-1]: # If both top and right pixels have the same label
        label_mat[i][j] = label_mat[i][j-1]
      else:
        label_mat[i][j] = label_mat[i][j-1]
        x = label_mat[i-1][j]
        y = label_mat[i][j-1]
        equi_pairs.append((x,y))

print(equi_pairs)
plt.imshow(label_mat,cmap="gray")
plt.show()

equiv_classes=[] # To store equivalent labels
item = equi_pairs[0]
class1 = list(item)
equiv_classes.append(class1)
print(equiv_classes)
for i in range(1,len(equi_pairs)):
  item = equi_pairs[i]
  flag = 0
  for j in range(len(equiv_classes)):
    k = equiv_classes[j]
    if(item[0] not in k) and (item[1] in k):
      equiv_classes[j].append(item[0])
      flag = 1
      break
    if(item[0] in k) and (item[1] not in k):
      equiv_classes[j].append(item[1])
      flag = 1
      break
  if flag==0:
    equiv_classes.append(list(item))
print(equiv_classes)
# Creating final labelled component matrix
for i in  range(label_mat.shape[0]):
  for j in range(label_mat.shape[1]):
    for k in equiv_classes:
      if label_mat[i][j] in k:
        label_mat[i][j] = k[0]
plt.imshow(label_mat,cmap="gray")
plt.show()
````

**Histogram Visualization**


```
cv2.imread(path,flag) : Method for loading image from specified path

Arguments:

path : From where the image is to be fetched

flag: specifies the mode to read the image, 0: to read as a grayscale image, 1: to read as a color image

cv2.calcHist(images, channels, mask, histSize, range) : Method to calculate histogram of an image

images: list of images 

channels: list of channels used to calculate histograms

mask: calculates histogram for masked pixels only if a mask is provided

histSize: specifies the number of bins

range:range of pixel values
```




````python
import cv2
img = cv2.imread('/content/drive/MyDrive/low_contrast.png',0)
img2 = cv2.imread('/content/drive/MyDrive/high_contrast.jpg',0)
img3 = cv2.imread('/content/drive/MyDrive/photo1.png',0)
hist = cv2.calcHist([img],[0],None,[256],[0,256])
print(img2.shape)
# show the plotting graph of an image
plt.title("Low contrast image(washed out)")
plt.imshow(img,cmap='gray')
plt.show()
plt.plot(hist)
plt.title("Low contrast image Histogram")
plt.show()

hist2 = cv2.calcHist([img2],[0],None,[256],[0,256])
# show the plotting graph of an image
plt.title("High contrast image")
plt.imshow(img2,cmap='gray')
plt.show()
plt.plot(hist2)
plt.title("High contrast image Histogram")
plt.show()

hist3 = cv2.calcHist([img3],[0],None,[256],[0,256])
# show the plotting graph of an image
plt.title("Low contrast image(dark image)")
plt.imshow(img3,cmap='gray')
plt.show()
plt.plot(hist3)
plt.title("Low contrast image Histogram")
plt.show()
````
